## Описание данных о технологичесом процессе

**Схема технологического процесса (этапы)**:
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики
![gold_process_image](gold_process.png)

**У каждого этапа могут быть различные типы параметров:**
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

**Технологический процесс**:
- `Rougher feed` — исходное сырье
- `Rougher additions (или reagent additions)` — флотационные реагенты Xanthate, Sulphate, Depressant:
  - `Xanthate` — ксантогенат (промотер, или активатор флотации);
  - `Sulphate` — сульфат (на данном производстве сульфид натрия);
  - `Depressant` — депрессант (силикат натрия).
- `Rougher process` — флотация (англ. «грубый процесс»)
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

**Параметры этапов**:
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

## Метрики
    
**Расчёт эффективности:**
- $C$ — доля золота в концентрате после флотации/очистки;
- $F$ — доля золота в сырье/концентрате до флотации/очистки;
- $T$ — доля золота в отвальных хвостах после флотации/очистки.

$$Recovery = \frac{C\times(F-T)}{F\times(C-T)}\times100\%$$

---
    
**Метрика качества sMAPE:**
- $y_i$ — значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество;
- $\hat{y}_i$ — значение предсказания для объекта с порядковым номером i, например, в тестовой выборке;
- $N$ — количество объектов в выборке.   

$$sMAPE = \frac{1}{N}\sum\limits_{i=1}^N\frac{|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2}\times100\%$$
____
    
**Итоговая метрика качества sMAPE:**

$$sMAPE_{result} = 0.25\times sMAPE_{rougher} + 0.75\times sMAPE_{final}$$

## Расчет MAE для эффективности обогащения

Рассчитаем эффективность обогащения по формуле выше на этапе флотации `rougher.output.recovery`:

```python
# доля золота в концентрате после флотации
C = gold_train['rougher.output.concentrate_au']

# доля золота в сырье до флотации
F = gold_train['rougher.input.feed_au']

# доля золота в хвостах после флотации
T = gold_train['rougher.output.tail_au']

# рассчитаем эффективность обогащения 
gold_train['rougher.output.recovery_check'] = (C*(F - T) / (F*(C - T))) * 100

gold_train[['rougher.output.recovery', 
            'rougher.output.recovery_check']].head()
```

```python
# определим количество пропусков
gold_train[['rougher.output.recovery', 'rougher.output.recovery_check']].isna().sum()
```

Т.к. пропуски мешают расчету MAE, то создадим отдельный датафрейм без них. 

```python
gold_train_cleared = gold_train[['rougher.output.recovery', 
                                 'rougher.output.recovery_check']].dropna()

# расчитаем среднюю абсолютную ошибку
MAE = mean_absolute_error(gold_train_cleared['rougher.output.recovery'],
                    gold_train_cleared['rougher.output.recovery_check'])

md('**Ошибка равна {:.16f}, т.е. очень близка к нулю, значит эффективность \
обогащения посчитана правильно.**'.format(MAE))
```

Удалим проверочный столбец `recovery_check`.

```python
gold_train.drop('rougher.output.recovery_check', axis=1, inplace=True)
```

## Предобработка

```python
def show_me_nan(data,columns):
    # зеленым цветом (#66ff00) - непропущенные, красным цветом (#d41243) - пропущенные данные 
    colors = ['#66ff00', '#d41243'] 
    fig, ax = plt.subplots(figsize=(10,5)) 
    sns.heatmap(data[columns].isna(), cmap=sns.color_palette(colors), 
                cbar=False, ax=ax)
    plt.xlabel('Название столбца в датафрейме', fontsize=16)
    plt.ylabel('Номер строки', fontsize=14)
    plt.title('Тепловая карта пропусков (пропуски красным цветом)', fontsize=14)
    plt.show()
```

Посмотрим на карту пропусков. Т.к. признаков много, то сначала первую сначала первую половину, затем вторую.

```python
show_me_nan(data=gold_train, columns=gold_train.columns[1:30])
```

```python
show_me_nan(data=gold_train, columns=gold_train.columns[30:55])
```

# Разработка модели

Выделим целевые признаки: эффективность обогащения чернового концентрата `rougher.output.recovery` и финального концентрата `final.output.recovery`.

```python
# удалим также дату как несущественный признак
features_train = gold_train.drop(['rougher.output.recovery', 
                                  'final.output.recovery', 'date'], axis=1)

target_train = gold_train[['rougher.output.recovery', 
                           'final.output.recovery']]
```

Используем кастомную кросс-валидацию:

```python
def sMAPE_calc(y, y_pred):
    """Считает метрику по требованиям заказчика"""
    sMAPE = np.mean((abs(y - y_pred) / ((abs(y) + abs(y_pred)) / 2))) * 100 
    return sMAPE
```

```python
# делаем кастомную кросс-валидацию.
def score_model(model, features, target, cv):
    scores = []
    for subtrain_index, valid_index in KFold(n_splits=cv).split(features):
        features_subtrain = features.loc[subtrain_index].reset_index(drop=True)
        target_subtrain = target.loc[subtrain_index].reset_index(drop=True)
        
        features_valid = features.loc[valid_index].reset_index(drop=True)
        target_valid = target.loc[valid_index].reset_index(drop=True)
        
        model.fit(features_subtrain, target_subtrain)
        
        # получаем предсказания: 
        # в первом столбце для rougher.output.recovery, во втором - для final.output.recovery
        pred_valid = pd.DataFrame(model.predict(features_valid))
        
        # используем свою метрику как функцию выше для этапов rougher и final
        scores_rougher = sMAPE_calc(target_valid.loc[:, 'rougher.output.recovery'], 
                                    pred_valid.loc[:, 0])
        
        scores_final = sMAPE_calc(target_valid.loc[:, 'final.output.recovery'], 
                                    pred_valid.loc[:, 1])
        
        # итоговая метрика
        scores.append(0.25*scores_rougher + 0.75*scores_final)
    
        # добавим в функцию построение распределений (истина и предсказания)
        target_median = target_valid.loc[:, 'rougher.output.recovery'].median()
        pred_median = pred_valid.loc[:, 0].median()
        
        #print(model.coef_)
        
        plt.figure(figsize = (11,5))
        g = sns.distplot(target_valid.loc[:, 'rougher.output.recovery'], 
                         color='blue', bins=100, label='истинные значения')
        g = sns.distplot(pred_valid.loc[:, 0], 
                         color='green', bins=100, label='предсказанные значения')
        #g.set_xlabel('rougher.input.feed_size', fontsize=16)
        g.set_title('Распределения для разных блоков при кросс-валидации', fontsize=16)
        plt.axvline(target_median, color='blue', linestyle='--')
        plt.axvline(pred_median, color='green', linestyle='--')
        g.set(xlim=(60, 110))
        plt.legend(fontsize=16)
        plt.show()
    
    return pd.Series(scores).round(2)
```

## Простая линейная регрессия

```python
lr_scores = score_model(LinearRegression(), features_train, target_train, cv=5)
print('Итоговые метрики sMAPE для линейной регрессии: ')
display(lr_scores)
md('**Средняя sMAPE для простой линейной регрессии: {:.3f} %**'.format(lr_scores.mean()))
```

```python
lr_scores = score_model(Lasso(), features_train, target_train, cv=5)
print('Итоговые метрики sMAPE для линейной регрессии с Lasso регуляризацией: ')
display(lr_scores)
md('**Средняя sMAPE для линейной регрессии с Lasso регуляризацией: {:.3f} %**'.format(lr_scores.mean()))
```

## Линейная регрессия с Ridge регуляризацией

```python
lr_scores = score_model(Ridge(), features_train, target_train, cv=5)
print('Итоговые метрики sMAPE для линейной регрессии с Ridge регуляризацией: ')
display(lr_scores)
md('**Средняя sMAPE для линейной регрессии с Ridge регуляризацией: {:.3f} %**'.format(lr_scores.mean()))
```

## Случайный лес
