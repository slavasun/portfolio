## Обработка пропусков

```python
def nan_check(data, columns):
    '''Создаёт тепловую карту пропусков для общей визуализации их количества,
    если признаков так много, что не помещаются на один рисунок, то следует 
    подавать на вход датафрейм по частям (столбцам)'''
    # зеленым цветом (#66ff00) - непропущенные, красным цветом (#d41243) - пропущенные данные 
    colors = ['#66ff00', '#d41243'] 
    fig, ax = plt.subplots(figsize=(10,4)) 
    sns.heatmap(data[columns].isna(), cmap=sns.color_palette(colors), cbar=False, ax=ax)
    #plt.xlabel('Название столбца в датафрейме', fontsize=16)
    plt.ylabel('Номер строки', fontsize=14)
    plt.title('Тепловая карта пропусков (пропуски красным цветом)', fontsize=14)
    plt.show()
    
    print('На наличие NaN значений в относительном выражении в %:\n',
          data.isna().mean())
    
    print('На наличие пропусков как строковых значений в шт.:\n', 
        data.isin([' ', '', '-', 'None', 'Null', 'null', 'NULL', 
        'NA', 'NaN', '-NaN', '-nan', 'nan', 'N/A', 'n/a', '#N/A', 
        '#N/A N/A', '#NA', '-1.#IND', '-1.#QNAN',  '1.#IND', '1.#QNAN']).sum())
  
    
    display(data.head().style.highlight_null())
```

```python
nan_check(data=df, columns=df.columns)
```

- Много пропусков в `VehicleType`, `Gearbox`, `Model`, `FuelType`, `NotRepaired`, чем заполнять пока непонятно, поэтому просто используем значение-пустышку -> 'unknown'

## Обработка выбросов (категориальные признаки)
К ним относятся: `VehicleType`, `Gearbox`, `Model`, `FuelType`, `Brand`, `NotRepaired`.

```python
def quick_count_box_plots(df, feature, label_angle=0):
    """Функцию применяем для категориальных переменных, строит 
    сразу два графика countplot и boxplot"""
    plt.figure(figsize=(12, 10))
    plt.subplot(2, 1, 1)
    ax = sns.boxplot(x=feature, y="Price", data=df, showfliers=True)
    ax.set_xlabel('')
    ax.set_xticklabels(ax.get_xticklabels(),rotation=label_angle)
    plt.title('Разброс цены автомобиля в зависимости от признака {}'.
              format(feature), fontsize=16)
    
    plt.subplots_adjust(hspace=0.4) 
    
    plt.subplot(2, 1, 2)
    ax = sns.countplot(df[feature])
    ax.set_xlabel('')
    ax.set_xticklabels(ax.get_xticklabels(),rotation=label_angle)
    plt.title('Количество записей для каждого значения признака {}'.
              format(feature), fontsize=16)
    plt.show()
```

# Построение моделей (models)
Еще раз, заказчику важны:
- качество предсказания как метрика RMSE;
- скорость предсказания на test выборке;
- время обучения на train выборке.

Все измерения по каждой модели сведем в итоговую таблицу в конце раздела.

```python
def my_rmse(y_true, y_predict):
    """Возвращает значение RMSE метрики"""
    mse = mean_squared_error(y_true, y_predict)
    rmse = np.sqrt(mse)
    return rmse
```

```python
def distplot_trueVSpredict(y_true, y_predict):
    """Выводит гистограммы истинных и предсказанных значений целевой переменной
    на одном графике"""
    plt.figure(figsize = (11,5))
    g = sns.distplot(y_true, color='green', bins=50, label='истинные значения')
    g = sns.distplot(y_predict, color='red', bins=50, label='предсказанные значения')
    g.set_title('Распределения значений цены автомобиля', fontsize=16)
    plt.legend(fontsize=16)
    plt.show()
```

## DummyMedianModel

```python
# создадим массив предсказаний с медианным значением
y_predict_median = pd.Series(y_train.median(), index=range(y_test.shape[0]))
md('**Медианное значение цены на обучающей выборке: {:.2f} евро.**'.format(y_train.median()))
```

```python
md('**RMSE дефолтной модели с медианой: {:.2f} евро**'.format(my_rmse(y_test, y_predict_median)))
```
